```python
def parse_rating(rating_classes):
    
    rating_map = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }
    
    for cls in rating_classes:
        if cls in rating_map:
            return rating_map[cls]
    
    return 0
```

```python
rating_classes = books_html[0].find("p", class_="star-rating")["class"]

parse_rating(rating_classes)
```

```python
book_rating = parse_rating(books_html[0].find("p", class_="star-rating").attrs['class'])
```

```python
books_dict = {
    "Title": [],
    "Price": [],
    "Rating": []
}
```

```python
import re

for book in books_html:
    
    # Title
    title = book.h3.a["title"]
    
    # Price
    price_text = book.find("p", class_="price_color").text
    price = float(re.sub(r"[^\d.]", "", price_text))
    
    # Rating
    rating_classes = book.find("p", class_="star-rating")["class"]
    rating = parse_rating(rating_classes)
    
    # Dictionary'ye ekle
    books_dict["Title"].append(title)
    books_dict["Price"].append(price)
    books_dict["Rating"].append(rating)
```

```python
df = pd.DataFrame(books_dict)
df.head()
```

```python
df.groupby("Rating").count()["Title"].plot(kind="bar")
```
